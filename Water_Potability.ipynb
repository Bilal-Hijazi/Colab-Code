```python
data = pd.read_csv("water_potability.csv")
import pandas as pd
from imblearn.over_sampling import SMOTE

# Assuming you've already loaded your dataset into a DataFrame called data
# Check class distribution
print("Class distribution before SMOTE:")
print(data['Potability'].value_counts())

data = data.dropna()
# Separate features and target variable
X = data.drop('Potability', axis=1)
y = data['Potability']

# Apply SMOTE
smote = SMOTE()
X_resampled, y_resampled = smote.fit_resample(X, y)

# Convert the resampled data back to a DataFrame
data_resampled = pd.concat([pd.DataFrame(X_resampled), pd.DataFrame({'Potability': y_resampled})], axis=1)

# Check class distribution after SMOTE
print("\nClass distribution after SMOTE:")
print(data_resampled['Potability'].value_counts())
data.head(15)
```
